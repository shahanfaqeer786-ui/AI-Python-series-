```python
## Datatype in csv
data1= ('a,b,c,d\n'
        '1,2,3,4\n'
        '5,6,7,8\n'
        '9,10,11')
```

```python
data_f1 = pd.read_csv(StringIO(data1),dtype={'a':int,'b':float,'c':'str','d':'object'})
data_f1.info()
```

```python
## Check dtypes
data_f1.dtypes
```

```python
data2=('index,a,b,c\n'
        '4,apple,bat,5.7\n'
        '8,orange,cow,10')
```

```python
## Use Index cols and usecols
data3=('a,b,c\n'
        '4,apple,bat,\n'
        '8,orange,cow,')
```
