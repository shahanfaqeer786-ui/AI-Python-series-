## What is XML?


1.   XML stands for eXtensible Markup Language.
2.   XML is a Markup Language much like HTML.

1.   XML was designed to store and transport data.
2.   XML was designed to be self-descriptive .

1.   XML is W3C Recommendation











```python
import pandas as pd
```

```python
pd.read_xml("test.xml.txt")
```

```python
xml = '''<?xml version='1.0' encoding='utf-8'?>
<data xmlns="http://example.com">
  <row>
    <shape>square</shape>
    <degree>360</degree>
    <sides>4.0</sides>
    <firstname>abuzar</firstname>
  </row>

  <row>
    <shape>circle</shape>
    <degree>360</degree>
    <sides></sides>
    <firstname/>
  </row>

  <row>
    <shape>triangle</shape>
    <degree>180</degree>
    <sides>3.0</sides>
    <firstname/>
  </row>
</data>
'''
```

```python
pd.read_xml(xml)
```

```python
## Second methon
xml1 = '''<?xml version='1.0' encoding='utf-8'?>
<data>
  <row shape="square" degree="360" sides = "4.0" firstname = "abuzar" />
  <row shape="circle" degree="360"/>
  <row shape="triangle" degree="180" sides = "3.0"  lastname="ali"/>
</data>
'''
```

```python
pd.read_xml(xml1)
```

```python
##Xpath Parameter
pd.read_xml(xml1,xpath = ".//row")
```

```python
## Namespaces
xml2 = '''<?xml version='1.0' encoding='utf-8'?>
<doc:data xmlns:doc="https://example.com">
  <doc:row>
    <doc:shape>square</doc:shape>
    <doc:degree>360</doc:degree>
    <doc:sides>4.0</doc:sides>
    </doc:row>

  <doc:row>
    <doc:shape>circle</doc:shape>
    <doc:degree>360</doc:degree>
    <doc:sides/>
    </doc:row>

 <doc:row>
 <doc:shape> triangle</doc:shape>
 <doc:degree>180</doc:degree>
 <doc:sides>3.0</doc:sides>
 </doc:row>
 </doc:data>
'''
```

```python
df=pd.read_xml(xml2,xpath=".//doc:row",namespaces={"doc":"https://example.com"})
df
```

```python
##Convert dataframe into xml file
df.to_xml("test1.xml")
```
