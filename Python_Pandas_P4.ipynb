```python
import pandas as pd
import requests

url = "https://en.wikipedia.org/wiki/Mobile_country_code"

# Add a browser-like User-Agent header
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/123.0.0.0 Safari/537.36"
}

# Fetch the HTML content manually
response = requests.get(url, headers=headers)

# Parse all tables in the HTML
html = pd.read_html(response.text)

# Example: show number of tables and the first few rows of the first table
# print(len(tables))
print(html)
```

```python
url2 = ("https://en.wikipedia.org/wiki/Minnesota")
response = requests.get(url2, headers=headers)
html2 = pd.read_html(response.text, match ="Average daily maximum and minimum temperatures for selected cities in Minnesota")
html2
```
