## Python Pandas Working With JSON Part03
In this Part we are going to learn about:


*   read json (read_json)
*   to json (to_json)


*   Json normalized






```python
data = '{"employee_name":"james","email":"james@gmail.com", "job_profile": [{"title1":"taem lead","title2":"sr.developer"}]}'
```

```python
type(data)
```

```python
import pandas as pd
```

```python
## Convert json data into dataframe
pd.read_json(data)
```

```python
## Orient in record by default it is orientated in record data

pd.read_json(data,orient='records')
```

```python
## Orient in index it will conver columns name in rows indexes
pd.read_json(data, orient = "index")
```

```python
## assignment =  split format k liye konsa data format hoga
pd.read_json(data, orient = "split")
```

```python
pd.read_json(data, orient = "column")
```

```python
df = pd.DataFrame([['a','b'], ['c','d']],
                   index=['Row1','Row2'],
                   columns=['col1','col2'])
df
```

```python
## Convert dataframe into json
j=df.to_json()
j
```

```python
df.to_json(orient='index')
```

```python
df.to_json(orient='columns')
```

```python
## orient record function will remove rows
df.to_json(orient='records')
```

```python
## split function will give record indiviully
df.to_json(orient='split')
```

```python
df.to_json(orient='table')
```

```python
## Contvert table form into dataFrame
schema = '{"schema":{"fields":[{"name":"index","type":"string"},{"name":"col1","type":"string"},{"name":"col2","type":"string"}],"primaryKey":["index"],"pandas_version":"1.4.0"},"data":[{"index":"Row1","col1":"a","col2":"b"},{"index":"Row2","col1":"c","col2":"d"}]}'
pd.read_json(schema, orient='table')
```

```python
## COnvert csv file into DataFrame
df=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data',header=None)
df.head()
```

```python
##COnvert csv file into json
df.to_json(orient="index")
```

```python
##COnvert csv file into json
df.to_json(orient="table")
```
